```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('web_clients_correct.csv')
df.head()
```

```python
# scatterplot через matplotlib
plt.scatter(df['age'], df['bill'])
```

```python
# scatterplot через seaborn
sns.scatterplot(data=df, x='age', y='bill')
```

```python
# regplot
sns.jointplot(x='age', y='bill', data=df, kind='reg')
```
